```python
import pandas as pd
df = pd.read_excel(r'Customer Call List.xlsx')
df
```

```python
df = df.drop_duplicates()
#remove duplicated raws
#df = is to save the opiration
```

```python
df = df.drop(columns= 'Not_Useful_Column' )
#remove column
```

```python
#df['Last_Name'] = df['Last_Name'].str.strip("...")
#df['Last_Name'] = df['Last_Name'].str.strip("/")
#df['Last_Name'] = df['Last_Name'].str.strip("_")
#df['Last_Name']
df['Last_Name'] = df['Last_Name'].str.strip("_.../")
df
#strip remove string from column lstrip remove from left, rstrip remove from right and strip remove from both
#df[] = df[] to save  the value of column
#you can remove all the redundancies by put them at the same time
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace( '[^a-zA-Z0-9]' ,'', regex= True) 
# str.replace('[^a-zA-Z0-9]','')means replace every thing with ''(nothing) except [^a-zA-Z0-9]
#regex is false by defult 
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x) )
#convert all values of the column to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:9] )
#separate the phone number by -
df
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].replace('Na--','')
df
```

```python
df[['Street Address' , 'State','Zip code']] = df['Address'].str.split(',' ,  n=2 , expand= True)
df
#split the column according to the first value then n is the number of spliting according to the same value
# expand is converting the split function into columns
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df
#repalces all yes and no with Y and N
```

```python
df = df.replace('N/a','')
df
#remove 'N/a'
```

```python
df = df.drop(columns= 'Address' )
#remove address column
df
```

```python
df = df.fillna('')
df
#remove all NaN values 
```

```python
for x in df.index:
    if df.loc[x , 'Do_Not_Contact'] == 'Y':
        df.drop(x,inplace=True)
df
#remove all rows that have Y in Do_Not_Contact column
```

```python
for x in df.index:
    if df.loc[x,'Phone_Number'] == '':
        df.drop(x,inplace=True)
df
#remove all rows with no phone number
#df.dropna(subset='Phone_Number' , inplace= True)  #another way to drop null values 
```

```python
df = df.reset_index(drop=True)
df
#reset the number of index and drop the previous index
```
