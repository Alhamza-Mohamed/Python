# Syntax and built in functions

## List Comprehension 
provides a concise and readable way to create a new list based on the values of existing iterable 

```python
words = ["This", "is", "An", "Example", "Of", "PYTHON", "syntax"]
word_lower = [word.lower() for word in words if len(word) > 3  ] # list comprehension
print(word_lower)
```

General form: [expression for item in iterable if condition] 

expression: what you want to produce for the new list. Can be a transformation of item (item.lower(), f"{item} {other}")






## Iterables 

An object capable of returning its members one at a time. Examples of iterables include all sequence types (such as list, str, and tuple) and some non-sequence types like dict, file objects


### Enumerate
built in function, its input is iterable(ex: list), output an iterator of (index, item) pairs

It can have argument called start which is 0 by default to initialize the start index 


```python
words = ["NLP", "is", "fun"] 
for index, value in enumerate(words):
    print (index, value) 
```

```python
for index, value in enumerate(words, start=1):
    print (index, value) 
```

### Zip

built in function, input 2 or more iterables (list, tuples, etc.), output an iterator of tuples, where each tuple contains items from the input iterables at the same position

```python
words = ["deep", "learning", "rocks"]
lengths = [4, 8, 5]

for item1,item2 in zip(words, lengths):
    print (item1, item2)
```

```python
words = ["deep", "learning", "rocks", "AI"]
lengths = [4, 8, 5, 2]

for item1,item2 in zip(words,lengths):
     if len(item1) > 4:
        print(item1, item2)


```

```python
words = ["deep", "learning", "rocks", "AI"]
lengths = [4, 8, 5, 2]
words_list = [f"{item1} {item2}"  for item1,item2 in zip(words,lengths) if len(item1) > 4 ]
print(words_list)

```

```python
tokens = ["nlp", "is", "fun", "and", "useful"]
freqs  = [5, 1, 4, 2, 3]

words_list = [f"{item1}:{item2}" for item1,item2 in zip(tokens,freqs) if item2 > 2]
print(words_list)
```

## Dictionary 

### 1- Form a list of pairs (Tuples)
If your list already contains pairs (like tuples or sub-lists), you can simply use the dict() constructor. This is the cleanest and fastest method.

```python
data_list = [("name", "Alice"), ("age", 30), ("city", "Giza")]
result = dict(data_list)
print(result)
```

```python
keys = ["a","b","c"]
result = {}

i = 0
for key in keys:
    result[key] = i
    i = i +1 
print (result)
```

### 2- Using dictionary comprehension
most flexible method.it allows you to transform data as you map it

- mapping to a constant value

```python
keys = ["a","b","c"]
i = 0
result = {key:i for key in keys  }
print(result)
```

- using the list index as the value

```python
items = ["apple", "banana", "cherry"]
result = {item: i for i, item in enumerate(items)}
print (result)
```

### 3- Paring Two lists (keys and values)
if you have 2 separete lists - one for keys and one for values - use zip() inside the dict() constructor

```python
keys = ["ID", "Status", "Type"]
values = [101, "Active", "Admin"]

result = dict(zip(keys, values))
print (result)
```

### 4- handling repeated keys
If your list has duplicate items and you want to count them or group them, use collections.defaultdict or the .get() method.

**Using a loop to count occurrences**

```python
fruits = ["apple", "orange", "apple", "pear", "orange", "apple"]
counts = {}

for fruit in fruits:
    counts[fruit] = counts.get (fruit, 0 ) + 1 
print(counts)
```

**Using collection counter**

```python

from collections import Counter

my_list = ["apple", "banana", "apple", "orange", "banana", "apple"]
counts = Counter(my_list)

print(counts)
```
